### The code in Cell 5 imports Numpy, defines two 2x2 arrays, multiplies these arrays using matrix multiplication and then prints the resulting matrix.

```python
import numpy as np

a = np.array([[4, 2], [7, 4]]);
b = np.array([[3, 6], [5, 8]]);

c = np.matmul(a, b);  # Matrix multiplication of a and b.

print(c);
```
